### Посмотрю на разнообразие ответов в колонках типа `object`

```python
# Фильтрую колонки по типу object
object_columns = df.select_dtypes(include=['object']).columns

# Выведу все уникальные значения для каждой колонки, чтобы понять данные
# Задаю параметр start = 1, тк помню, что колонка с индексом [0] - данные типа int - порядковый номер записи
for i, column in enumerate(object_columns, start = 1):
    print(f"Колонка {i}: '{column}'")
    print(f"Уникальные значения: {df[column].unique()} \n")
    print('-'*66)
```

### Разнообразие ответов на вопросы дало мне понимание, что для удобства их интерпретации мне надо привести всё в один вид: из типа данных `object` необходимо првиести к типу данных `int` по следующим правилам: 
- Колонка NPS будет допускать только множество ответов `{[1; 10], [99]}`, где 1 - плохая оценка, 10 - отличная, 99 - не учитывается.
- Колонки с оценками будут допускать множество ответов `{[1; 5]}`, где 1 - плохая оценка, 5 - отличная.

Для этого воспользуюсь удобным инструментом `extract` из pandas, где регулярное выражение (reg.exp) `\d+` извлекает первую последовательность цифр из каждой строки в колонке, а затем приведу всё к типу `int` модулем `astype`

```python
# Для каждой колонки типа object
for column in object_columns:
    # Оставляю лишь первую цифру в каждой строке
    df[column] = df[column].str.extract('(\d+)').astype(int)
```

### Расчёт индекса NPS

**Коротко о результате:** 
> `NPS равен 50%`, что указывает на хорошую степень лояльности и удовлетворённости клиентов. Это означает, что промоутеров больше чем хейтеров, но есть место для улучшений.**

**Воспользуюсь следующим правилом расчёта:**

<img src="2.png" width=500>
<img src="1.png" width=500>

```python
# Выберу колонку для расчёта NPS
column_nps = df.columns[1]

# Подсчёт количества ячеек с диапазоном от 0 до 6 включительно
haters = df[(df[column_nps] >= 0) & (df[column_nps] <= 6)].shape[0]

# Подсчёт количества ячеек с диапазоном от 9 до 10 включительно
promoters = df[(df[column_nps] >= 9) & (df[column_nps] <= 10)].shape[0]

# Подсчёт количества всех остальных ячеек
other = df.shape[0] - haters - promoters

# Подсчёт общего количества ячеек в колонке
total = df.shape[0]

# Вывод результатов
print(f"Критиков: {haters}")
print(f"Промоутеров: {promoters}")
print(f"Прочее: {other}")
print(f"Общее количество ячеек в колонке: {total}")
```

```python
NPS = (promoters / total - haters / total) * 100
print(f"NPS равен {NPS} %")
```

## 2. Создать пременную с перекодировками возраста. У нас есть переменная Age (лист База данных). Необходимо создать переменную (можно формулами в ексель или кодом - как удобно) возрастные группы, то есть, например, нужно сделать интервалы 18-25лет, 26-35лет, 36лет и старше
### То есть, если человек вписал в Age, что ему 19лет, то он в новой переменной будет в группе 18-25 лет. Если человек вписал в Age, что ему 40лет, то он в новой переменной будет в группе "36лет и старше" и так далее

**1 способ: тривиальный и нативный**:

```python
# Функция для определения возрастной группы
def get_age_group(age):
    if age <= 25:
        return '18-25 лет'
    elif age <= 35:
        return '26-35 лет'
    else:
        return '36 лет и старше'

# Создание переменной возрастных групп
df['groups'] = df['Age'].apply(get_age_group)
```

Здесь я написал функцию, которая принимает значение возраста и возвращает соответствующую категорию строчкой - заполняет этой строкой соответствующую ячейку. Метод `apply()` применяет эту функцию к КАЖДОМУ значению в столбце Age и сохраняет результат в новом столбце groups в датасете.

**Проверим результат**

```python
shrek = df.iloc[:, [0, -2, -1]].sample(5)
shrek
```

**2 способ: в pandas есть для этого метод `cut` и ниже приведена его документация, интереса ради надо попробовать!**

Я помню, что apply не предпочтительный способ применения функции к данным, потому что он сильно замедляет процесс при огромном массиве данных.

```python
# Создание интервалов возраста
bins = [18, 25, 35, float('inf')]

# Создание категорий возраста
labels = ['18-25 лет', '26-35 лет', '36 лет и старше']

# Создание переменной возрастных групп
df['age_groups'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
```

## 3. Построение регрессионной модели.
### 3.1 Какие именно характеристики сильнее всего влияют на показатель NPS?
### 3.2 Интерпретация R-квадрата.

Коротко о результатах:
`R-квадрат = 0.305`
- это означает, что 30.5% изменчивости зависимой переменной NPS может быть объяснено независимыми переменными - результатами опроса и возрастом, включенными в модель. 
- Это означает, что модель объясняет лишь часть изменчивости зависимой переменной, и оставшиеся 69.5% изменчивости могут быть вызваны другими факторами, которые не учтены в модели.

Еще вероятно некоторые признаки кореллируют друг с другом и не выполнено условие мультиколлинеарности признаков, также вероятно на бОльшем количестве данных модель сыграла бы по-другому.

**Уровень лояльности потребителей в контексте данного опроса зависит от двух главных фаткоров: отсутствии навязанных услуг и способности консультантов решать проблемы. В целом это все говорит о том, что главное для клиентов - не испытывать лишних проблем, а при их возникновении они должны быстро и легко решаться за счет консультантов.** 

```python
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Выбор независимых переменных - колонок результатов опроса и колонки возраста
X = df.iloc[:, 2:22]

# Выбор целевой переменной - колонки показателя NPS
y = df.iloc[:, 1]
```

Добавление константы в матрицу независимых переменных необходимо для учета свободного члена (intercept) в линейной регрессионной модели. Это позволяет модели лучше аппроксимировать данные и улучшить ее качество. 

Изучал правильность построения модели исходя из источников: 
> https://robotdreams.cc/blog/437-shcho-take-liniyna-regresiya

> https://education.yandex.ru/handbook/data-analysis/article/linejnaya-regressiya-metod-naimenshih-kvadratov

```python
# Добавление константы в матрицу независимых переменных
X = sm.add_constant(X)
```

Разделение данных на обучающую и тестовую выборки
- 80% данных будут использоваться для обучения модели, а оставшиеся 20% - для тестирования модели.

```python
train_X = X.sample(frac=0.8, random_state=1)
train_y = y.loc[train_X.index]
test_X = X.drop(train_X.index)
test_y = y.drop(train_X.index)
```

Построение линейной регрессионной модели

```python
model = sm.OLS(train_y, train_X)
results = model.fit()
```

Вывод результатов модели

```python
print(results.summary())
```
